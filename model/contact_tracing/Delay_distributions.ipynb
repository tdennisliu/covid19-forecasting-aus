```python
# Delay distributions

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
plt.style.use("seaborn-poster")
```

```python
# DELAY 1 

fig,ax = plt.subplots(figsize=(12,9))

#t_p_shape = 1.0045815
#t_p_scale = 1/0.3292532
t_p_shape = 1.0310793
t_p_scale = 1/0.4150334
t_p_offset = 0  

p = t_p_offset + np.random.gamma(t_p_shape, t_p_scale,size=10000)
p = [x for x in p if x<14]
p_90 = np.percentile(p,90)


print("Median: %f.2" %np.median(p))
print("Mean: %f.2" %np.mean(p))
print("Variance: %f.2" %np.var(p))
print("90th Percentile: %f.2" %np.percentile(p,90))
print("98th Percentile: %f.2" %np.percentile(p,98))
ax.hist(p,bins=40, alpha=0.4, density=True)
ax.set_title("Delay 1: symptoms to swab distribution")
ax.set_xlabel("Days")
plt.savefig("figs/delay1.png",dpi=120)
plt.show()
```

```python
# CURRENT DELAY DISTRIBUTION (3/2,2)

fig,ax = plt.subplots(figsize=(12,9))

x = 1+ np.random.gamma(2, 1, size = 10000)

print("Mean: %f.2" %np.mean(x))
print("Variance: %f.2" %np.var(x))
#ax.hist(x,bins=40)
ax.set_title("Time to action distribution")

truncate=True
t_p_shape = 1.0310793
t_p_scale = 1/0.4150334
t_p_offset = 0     

t_t_shape = 7.905380
t_t_scale = 1/4.857156
t_t_offset =0

t_n_shape = 0.4533819
t_n_scale = 1/1.8199947
t_n_offset = 0

t_c_shape = 0.4533819
t_c_scale = 1/1.8199947
t_c_offset = 0


p = t_p_offset + 1*np.random.gamma(t_p_shape, t_p_scale,size=10000)

counter = 0
while truncate:
    p = [x for x in p if x<14]
    if counter>20:
        break
    p.extend(t_p_offset + np.random.gamma(t_p_shape,t_p_scale,size=10000-len(p)))
    counter +=1
    if np.all(np.array(p)<7):
        break

t = t_t_offset + 0.85*np.random.gamma(t_t_shape, t_t_scale, size = 10000)

n = t_n_offset + np.random.gamma(t_n_shape, t_n_scale, size = 10000)

c = t_c_offset + np.random.gamma(t_c_shape, t_c_scale, size = 10000)

ax.hist(p+t+n+c, bins=40, alpha=0.4,density=True)
ax.set_xlabel("Days")
plt.savefig("figs/total3_delays.png",dpi=120)
plt.show()


print("Mean: %f.2" %np.mean(p+t+n))
print("Median: %f.2" %np.percentile(p+t+n,50))
print("Variance: %f.2" %np.var(p+t+n))
print("90th Percentile: %f.2" %np.percentile(p+t+n,90))
```

```python
import numpy as np
import matplotlib.pyplot as plt

i=3.64
j=3.07
m=5.51
n=0.948
#Generation time distribution
fig,ax = plt.subplots(figsize=(12,9))
x = np.random.gamma(i/j, j, size = 10000)
print("Generation time: \nMean: %f.2" %np.mean(x))
print("Std: %f.2" %np.std(x))
ax.hist(x,bins=40, density=True)
ax.set_title("Generation time distribution")

#Symptom onset time distribution
fig,ax = plt.subplots(figsize=(12,9))
y =  np.random.gamma(m/n, n, size = 10000)
print("Symptom Onset: \nMean: %f.2" %np.mean(y))
print("Median: %f.2" %np.median(y))
print("Std: %f.2" %np.std(y))
ax.hist(y,bins=40)
ax.set_title("Symptom onset time distribution")


#difference 
fig,ax = plt.subplots(figsize=(12,9))

print("Difference \nMean: %f.2" %np.mean(x-y))
print("Variance: %f.2" %np.var(x-y))
ax.hist(x-y,bins=40)
ax.set_title("difference time distribution")
print("The 40 pc quantile of the TG - TS is %.2f" % np.quantile(x-y,0.4))
plt.show()
```
