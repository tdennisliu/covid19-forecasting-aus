```python
# Delay distributions

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# CURRENT DELAY DISTRIBUTION (3/2,2)

fig,ax = plt.subplots(figsize=(12,9))

x = 1+ np.random.gamma(2, 1, size = 10000)

print("Mean: %f.2" %np.mean(x))
print("Variance: %f.2" %np.var(x))
ax.hist(x,bins=40)
ax.set_title("Time to action distribution")

t_p_shape = 1
t_p_scale = 1
t_p_offset = 0.5   

t_t_shape = 1
t_t_scale = 1/2
t_t_offset =0.5

t_n_shape = 1
t_n_scale = 1/2
t_n_offset = 0

p = t_p_offset + np.random.gamma(t_p_scale, t_p_scale,size=10000)

t = t_t_offset + np.random.gamma(t_t_shape, t_t_scale, size = 10000)

n = t_n_offset + np.random.gamma(t_n_shape, t_n_scale, size = 10000)

ax.hist(p+t+n, bins=40, alpha=0.4)

plt.show()
```
