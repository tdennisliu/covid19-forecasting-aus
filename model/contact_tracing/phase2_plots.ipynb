```python
# select cases columns and columns to group by 

df_cases=df[['0obs_cases','1obs_cases','3obs_cases','4obs_cases','5obs_cases',
             '6obs_cases','7obs_cases','8obs_cases','9obs_cases','10obs_cases',
             '11obs_cases','12obs_cases','13obs_cases','14obs_cases','15obs_cases',
             '16obs_cases','17obs_cases','18obs_cases','19obs_cases','20obs_cases',
             '21obs_cases','22obs_cases','23obs_cases','24obs_cases','25obs_cases',
             '26obs_cases','27obs_cases','28obs_cases','29obs_cases',
            'detect']]

df_cases
```

```python
df_cases.columns = df_cases.columns.str[:-9]
df_cases.columns = [*df_cases.columns[:-1], 'detect']
df_cases
```

```python
# calculate median by group
m = df_cases.groupby('detect').median()
m
# transpose dataframe for plot
m_T = m.T
# remove first observation for plotting
m_T = m_T.iloc[1:]
# add day as variable and column names
m_T['Day'] = m_T.index
# convert Day to numeric
m_T["Day"] = pd.to_numeric(m_T["Day"])
m_T.columns = ['0.3', '0.5', '0.8', 'Day']
# melt dataframe for plotting 
m_T_long = pd.melt(m_T, id_vars=['Day'], value_vars=['0.3','0.5','0.8'],var_name='detection probability', value_name='Median observed cases')


# calculate lower 5% by group
# 5% interval
l_5 = df_cases.groupby('detect').quantile(0.05)
l_5_T = l_5.T
l_5_T = l_5_T.iloc[1:]
l_5_T['Day'] = l_5_T.index
l_5_T["Day"] = pd.to_numeric(l_5_T["Day"])
l_5_T.columns = ['0.3', '0.5', '0.8', 'Day']
l_5_T_long = pd.melt(l_5_T, id_vars=['Day'], value_vars=['0.3','0.5','0.8'],var_name='detection probability', value_name='lower 5')

# calculate upper 5% by group
u_5 = df_cases.groupby('detect').quantile(0.95)
u_5_T = u_5.T
u_5_T = u_5_T.iloc[1:]
u_5_T['Day'] = u_5_T.index
u_5_T["Day"] = pd.to_numeric(u_5_T["Day"])
u_5_T.columns = ['0.3', '0.5', '0.8', 'Day']
u_5_T_long = pd.melt(u_5_T, id_vars=['Day'], value_vars=['0.3','0.5','0.8'],var_name='detection probability', value_name='upper 5')


# calculate lower 25% by group
# 25% interval
l_25 = df_cases.groupby('detect').quantile(0.25)
l_25_T = l_25.T
l_25_T = l_25_T.iloc[1:]
l_25_T['Day'] = l_25_T.index
l_25_T["Day"] = pd.to_numeric(l_25_T["Day"])
l_25_T.columns = ['0.3', '0.5', '0.8', 'Day']
l_25_T_long = pd.melt(l_25_T, id_vars=['Day'], value_vars=['0.3','0.5','0.8'],var_name='detection probability', value_name='lower 25')


# calculate upper 25% by group
u_25 = df_cases.groupby('detect').quantile(0.75)
u_25_T = u_25.T
u_25_T = u_25_T.iloc[1:]
u_25_T['Day'] = u_25_T.index
u_25_T["Day"] = pd.to_numeric(u_25_T["Day"])
u_25_T.columns = ['0.3', '0.5', '0.8', 'Day']
u_25_T_long = pd.melt(u_25_T, id_vars=['Day'], value_vars=['0.3','0.5','0.8'],var_name='detection probability', value_name='upper 25')


all_long = m_T_long
all_long['lower 5'] = l_5_T_long['lower 5']
all_long['lower 25'] = l_25_T_long['lower 25']
all_long['upper 5'] = u_5_T_long['upper 5']
all_long['upper 25'] = u_25_T_long['upper 25']

#print(all_long)



```

```python
sns.set_style("white")
g = sns.FacetGrid(all_long, row="detection probability", height=3.5, aspect=3)
g.map(sns.lineplot, "Day", "Median observed cases", alpha=.7)
g.map(plt.fill_between,"Day","lower 25","upper 25",alpha=0.1,color='C0')
g.map(plt.fill_between,"Day","lower 5","upper 5",alpha=0.1,color='C0')
g.set(xlabel='Days since beginning of outbreak', ylabel='Number of observed cases')

print(g)
```
